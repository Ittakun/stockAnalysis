```python
# 基本パッケージ（numpy,Pandas,matplotlib）
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# globのimport
import glob
```

```python
# 表示関連
# DataFrameの列数設定
pd.set_option('display.max_columns', 100)
```

```python
# 2001～2018年の株価データをマージする
## globでファイル名の一覧を取得
stock_price_files = glob.glob('stockPrice/*.csv')
stock_price_list = []

# ファイルを読み込み、DataFrameでlistに格納する。
for f in stock_price_files:
    stock_price_list.append(pd.read_csv(f, header=1, encoding="shift-jis", parse_dates = [0]))
    
# Listに格納されたデータを全てconcat関数で連結    
stock_price_all = pd.concat(stock_price_list)

# 欠損値のチェック
print(stock_price_all.isnull().sum())

# 読み込みデータの表示
display(stock_price_all)
```

```python
# 2002～2018年の為替レートを取得
market_Data = pd.read_csv('market/market.csv', header=0, encoding="shift-jis", parse_dates = [0])

# 欠損値のチェック
print(market_Data.isnull().sum())

# レート情報をずらす
market_Data = market_Data.shift(-1)

# 読み込みデータの表示
display(market_Data)
```

```python
# 内部結合で結合する
merge_Data = pd.merge(stock_price_all, market_Data, on='日付', how='inner')

# 目的変数の作成
merge_Data['diff'] = merge_Data['終値'] - merge_Data['始値']
dummyData = pd.DataFrame({'answer' : merge_Data['diff'].apply( lambda x: 0 if x < 0 else 1 )})

# 月（1～12）、日（1～31）、曜日（月曜が0, 日曜が6）情報の抽出
dummyData = pd.DataFrame({
                          'month' : merge_Data['日付'].dt.month,
                          'day'  : merge_Data['日付'].dt.day,
                          'weekday' : merge_Data['日付'].dt.dayofweek
                         })

# 月、曜日情報をダミー変数へ変換
dummyData = pd.get_dummies(dummyData, columns=['month','day','weekday'])

# 最初の5行を表示
dummyData.head()
```

```python


# 不要な列を削除
merge_Data.drop(columns=['日付','終値','diff'], inplace=True)

# 各列を変化率へ一括変換
merge_Data_change = merge_Data.pct_change()

# 正解ラベルとダミー変数の付与
merge_Data_change = pd.concat([merge_Data_change, dummyData], axis=1)

# 正解ラベルをずらし、説明変数から見て未来の値とする
merge_Data_change.answer = merge_Data_change.answer.shift(-1)

# 最初と最後の行（NaNがある）を削除
merge_Data_change.drop(0, axis=0, inplace=True)
merge_Data_change.drop(len(merge_Data_change), axis=0, inplace=True)
```

```python
# ビニング
## 始値 10分割
bins = np.linspace(merge_Data_change['始値'].min(), merge_Data_change['始値'].max(), 10)
merge_Data_change['始値_bin'] = np.digitize(merge_Data_change['始値'], bins=bins)

## 安値 10分割
bins = np.linspace(merge_Data_change['安値'].min(), merge_Data_change['安値'].max(), 10)
merge_Data_change['安値_bin'] = np.digitize(merge_Data_change['安値'], bins=bins)

## 出来高 10分割
bins = np.linspace(merge_Data_change['出来高'].min(), merge_Data_change['出来高'].max(), 10)
merge_Data_change['出来高_bin'] = np.digitize(merge_Data_change['出来高'], bins=bins)

## 終値調整値 10分割
bins = np.linspace(merge_Data_change['終値調整値'].min(), merge_Data_change['終値調整値'].max(), 10)
merge_Data_change['終値調整値_bin'] = np.digitize(merge_Data_change['終値調整値'], bins=bins)

## 米ドル 10分割
bins = np.linspace(merge_Data_change['米ドル'].min(), merge_Data_change['米ドル'].max(), 10)
merge_Data_change['米ドル_bin'] = np.digitize(merge_Data_change['米ドル'], bins=bins)

## 英ポンド 10分割
bins = np.linspace(merge_Data_change['英ポンド'].min(), merge_Data_change['英ポンド'].max(), 10)
merge_Data_change['英ポンド_bin'] = np.digitize(merge_Data_change['英ポンド'], bins=bins)

## ユーロ 10分割
bins = np.linspace(merge_Data_change['ユーロ'].min(), merge_Data_change['ユーロ'].max(), 10)
merge_Data_change['ユーロ_bin'] = np.digitize(merge_Data_change['ユーロ'], bins=bins)

## カナダドル 10分割
bins = np.linspace(merge_Data_change['カナダドル'].min(), merge_Data_change['カナダドル'].max(), 10)
merge_Data_change['カナダドル_bin'] = np.digitize(merge_Data_change['カナダドル'], bins=bins)

## スイスフラン 10分割
bins = np.linspace(merge_Data_change['スイスフラン'].min(), merge_Data_change['スイスフラン'].max(), 10)
merge_Data_change['スイスフラン_bin'] = np.digitize(merge_Data_change['スイスフラン'], bins=bins)

## スウェーデンクローネ 10分割
bins = np.linspace(merge_Data_change['スウェーデンクローネ'].min(), merge_Data_change['スウェーデンクローネ'].max(), 10)
merge_Data_change['スウェーデンクローネ_bin'] = np.digitize(merge_Data_change['スウェーデンクローネ'], bins=bins)

## デンマーククローネ 10分割
bins = np.linspace(merge_Data_change['デンマーククローネ'].min(), merge_Data_change['デンマーククローネ'].max(), 10)
merge_Data_change['デンマーククローネ_bin'] = np.digitize(merge_Data_change['デンマーククローネ'], bins=bins)

## オーストラリアドル 10分割
bins = np.linspace(merge_Data_change['オーストラリアドル'].min(), merge_Data_change['オーストラリアドル'].max(), 10)
merge_Data_change['オーストラリアドル_bin'] = np.digitize(merge_Data_change['オーストラリアドル'], bins=bins)

## ニュージーランドドル 10分割
bins = np.linspace(merge_Data_change['ニュージーランドドル'].min(), merge_Data_change['ニュージーランドドル'].max(), 10)
merge_Data_change['ニュージーランドドル_bin'] = np.digitize(merge_Data_change['ニュージーランドドル'], bins=bins)

## 南アフリカランド 10分割
bins = np.linspace(merge_Data_change['南アフリカランド'].min(), merge_Data_change['南アフリカランド'].max(), 10)
merge_Data_change['南アフリカランド_bin'] = np.digitize(merge_Data_change['南アフリカランド'], bins=bins)

## バーレーンディナール 10分割
bins = np.linspace(merge_Data_change['バーレーンディナール'].min(), merge_Data_change['バーレーンディナール'].max(), 10)
merge_Data_change['バーレーンディナール_bin'] = np.digitize(merge_Data_change['バーレーンディナール'], bins=bins)

## インドルピー 10分割
bins = np.linspace(merge_Data_change['インドルピー'].min(), merge_Data_change['インドルピー'].max(), 10)
merge_Data_change['インドルピー_bin'] = np.digitize(merge_Data_change['インドルピー'], bins=bins)

## シンガポールドル10分割
bins = np.linspace(merge_Data_change['シンガポールドル'].min(), merge_Data_change['シンガポールドル'].max(), 10)
merge_Data_change['シンガポールドル_bin'] = np.digitize(merge_Data_change['シンガポールドル'], bins=bins)

## 韓国ウォン 10分割
bins = np.linspace(merge_Data_change['韓国ウォン'].min(), merge_Data_change['韓国ウォン'].max(), 10)
merge_Data_change['韓国ウォン_bin'] = np.digitize(merge_Data_change['韓国ウォン'], bins=bins)

## タイバーツ10分割
bins = np.linspace(merge_Data_change['タイバーツ'].min(), merge_Data_change['タイバーツ'].max(), 10)
merge_Data_change['タイバーツ_bin'] = np.digitize(merge_Data_change['タイバーツ'], bins=bins)

## クウェートディナール 10分割
bins = np.linspace(merge_Data_change['クウェートディナール'].min(), merge_Data_change['クウェートディナール'].max(), 10)
merge_Data_change['クウェートディナール_bin'] = np.digitize(merge_Data_change['クウェートディナール'], bins=bins)

## UAEディルハム 10分割
bins = np.linspace(merge_Data_change['UAEディルハム'].min(), merge_Data_change['UAEディルハム'].max(), 10)
merge_Data_change['UAEディルハム_bin'] = np.digitize(merge_Data_change['UAEディルハム'], bins=bins)

## メキシコペソ 10分割
bins = np.linspace(merge_Data_change['メキシコペソ'].min(), merge_Data_change['メキシコペソ'].max(), 10)
merge_Data_change['メキシコペソ_bin'] = np.digitize(merge_Data_change['メキシコペソ'], bins=bins)

# ダミー変数へ変換
merge_Data_change = pd.get_dummies(merge_Data_change, columns=['始値_bin','安値_bin','出来高_bin','終値調整値_bin','米ドル_bin',
                                                               '英ポンド_bin','ユーロ_bin','カナダドル_bin','スイスフラン_bin',
                                                              'スウェーデンクローネ_bin','デンマーククローネ_bin','オーストラリアドル_bin',
                                                              'ニュージーランドドル_bin','南アフリカランド_bin','バーレーンディナール_bin',
                                                              'インドルピー_bin','シンガポールドル_bin','韓国ウォン_bin','タイバーツ_bin',
                                                              'クウェートディナール_bin','UAEディルハム_bin','メキシコペソ_bin'])
```
